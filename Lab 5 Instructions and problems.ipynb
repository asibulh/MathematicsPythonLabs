## Problem 1

You toss a ball straight up at given velocity $v_0$. It leaves your hand at $y_0$ above a floor (See example at page 7 in Software 1, Orientation Physics lecture notes).

The position of the ball can be modeled with equation $y(t) = y_0 + v_0 t - \frac{1}{2}g t^2$, where

- $y_0$ is the initial position (m)
- $v_0$ is the initial velocity (m/s)
- $g$ is the acceleration due to gravity (9.81 m/s)
- $t$ is the time.

When does the ball hit the ground? Give the value to variable `t_end` with accuracy of 0.01 s. Print the value.

Draw the graph of the position of the ball over time. Use time step of 0.01 s. Stop the trajectory where the ball hits the ground.

What is the highest point of the ball? Store the value to variable `y_max` with two decimals accuracy in units of meters.

```python
# Initialize random number generator with student ID
rng = np.random.default_rng(student_id)
# Generate random initial conditions for projectile motion
v0 = rng.uniform(10, 20)  # Initial velocity in m/s
y0 = rng.uniform(1, 2)    # Initial height in m
print(f"Initial velocity (v0): {v0:.2f} m/s")
print(f"Initial height (y0): {y0:.2f} m")
```

```python
# Your code comes here
g = 9.81  # gravitational acceleration

# Solve quadratic equation:
# y(t) = y0 + v0*t - 0.5*g*t^2 = 0

coefficients = [-0.5 * g, v0, y0]

roots = np.roots(coefficients)

# Take only the positive root
t_end = np.max(roots)

print(f"Ball hits the ground after {t_end:.2f} seconds")

# Create time vector
t = np.linspace(0, t_end, 1000)

# Position of ball
y = y0 + v0 * t - 0.5 * g * t**2

# Plot
plt.figure(figsize=(6,4))
plt.plot(t, y)

plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.title("Ball motion")
plt.show()
```

## Problem 2

Your aim is to throw a ball to a target standing behind the fence (see the graph generated by the code). You stand on front of the fence. The ball leaves your hand $y_0$ meters above the ground. The following code simulates the situation. 

Your task is to find by trial and error **the minimum angle ($\theta$)**, **speed ($v_0$)** and **end of flight ($t_{end}$)** to hit the target (See also Software 1, Orientation Physics lecture notes, page 14).

Avoid to hit the fence.


```python
# Initial velocity and angle
v0 = 12.37 ## Change this !!!
angle = 31.1 ## Change this !!!
# End of flight
t_end = 1.22 ## Change this !!!

### DO NOT CHANGE ANYTHING BELOW THIS LINE ###

# Initialize random number generator with student ID
rng = np.random.default_rng(student_id)

# Initial position
x0 = 0.0
y0 = rng.uniform(0.5, 1.5)  # Initial height in m
print(f'x0 = {x0:.2f} m')
print(f'y0 = {y0:.2f} m')

theta = np.deg2rad(angle)
print(f'v0 = {v0:.2f} m/s')
print(f'angle = {angle:.2f} deg')
print(f't_end = {t_end:.2f} s')

g = 9.81 # Gravitational constant

# Define the initial conditions and parameters for the projectile motion
# This code simulates the trajectory of a ball thrown towards a target
yl = rng.uniform(2, 3)  # Level height in m
xl0 = rng.uniform(2, 5)  # Level x position in m
xw = rng.uniform(1, 3)  # Level x width in m

# Target position
x_target = rng.uniform(0, 5) + np.array([10, 9.5, 9.5, 10, 10])
y_target = rng.uniform(0, 1) + np.array([0.5, 0.5, 0.3, 0.3, 0.5])

# Fence position
x_edge = np.array([0, xl0, xl0, xl0+xw, xl0+xw, 20])
y_edge = np.array([0, 0,   yl,  yl, 0,   0])

# Define the time vector for the trajectory
t = np.arange(0, t_end, 0.001)

# Calculate the trajectory of the projectile
x = x0 + v0 * np.cos(theta) * t
y = y0 + v0 * np.sin(theta) * t - 0.5 * g * t**2

# Create figure
plt.figure(figsize=(6, 4))
# Plot the edge and level
plt.plot(x_edge, y_edge, 'b--', label = 'Fence')
plt.plot(x_target, y_target, 'r-', label = 'Target')
# plot the trajectory
plt.plot(x0, y0, 'g^', label = 'Start')
plt.plot(x, y, 'g-', label = 'Trajectory')
plt.plot(x[-1], y[-1], 'go', label = 'End')

#graph
plt.grid()
plt.xlabel('x position (m)')
plt.ylabel('y position (m)')
plt.title('Hit the target game')
plt.legend()
plt.xlim(-1, 21)
plt.ylim(-0.2, 7)
plt.show()
```
