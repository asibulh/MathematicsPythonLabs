## Problem 2

```python
# Run the code below to see your problem text
txt = mx.lab_2_problem_2(student_id)
display(Markdown(txt))

x = np.linspace(0, 10, 500)
k = 3.2
c = 5.40
f = np.sqrt(k * x)
g = (x - c)** 2
```

### My solution

```python
# Your code starts here

# Overview graph
plt.figure(figsize = (8,5))
plt.plot(x, f, label="f(x)")
plt.plot(x, g, label="g(x)")

plt.title("Overview: Both functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Intersection 01
plt.figure(figsize = (8,5))

plt.plot(x, f, label="f(x)")
plt.plot(x, g, label="g(x)")

plt.xlim(3.2, 4.2)     #adjust
plt.ylim(2, 5)
plt.title("Intersection Point 1 (zoomed)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()


#Intersection 02

plt.figure(figsize = (8,5))

plt.plot(x, f, label="f(x)")
plt.plot(x, g, label="g(x)")

plt.xlim(7, 8.2)     #adjust
plt.ylim(4, 8)
plt.title("Intersection Point 2 (zoomed)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

#x1 =  3.60  (approximately)
#x2 =  7.50  (approximately)
```
