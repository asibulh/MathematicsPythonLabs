### My solution

```python
# start your code here
#x values
x = np.linspace(-10, 10, 2000)

# piecewise function
y = (
((x>= 0.5) & (x< 3.5)) * (-0.5 + x) +
((x>= 3.5) & (x< 6.5)) * (6.5- x)
)
#plot
plt.figure(figsize=(7,4))
plt.plot(x, y, label="Triangle Function")
plt.title(f"Smooth Triangle Piecewise Function for student {student_id}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Run the code below to see your problem text
txt = mx.lab_3_problem_2(student_id)
display(Markdown(txt))
```

### My solution

```python
# Your code starts here
#x values
x = np.linspace(-10, 10, 2000)

# Piecewise function using boolean arrays
y = (x < 0) * (x**2) + (x >= 0) * (x)

#plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label="Piecewise Continuous Function")

plt.title(f"Smooth Continuous Piecewise Function for student {student_id}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
